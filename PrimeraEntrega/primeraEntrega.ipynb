## Exploratory Data Analysis (EDA)

(1) AB : This contains 25 .npy files consisting of a and b dimensions of LAB color space images, of the MIRFLICKR25k randomly sized colored image dataset. The LAB color space generally takes up large disk spaces, hence is a lot slower to load. That is the reason, I divided this into 25 files, so that it can be loaded at the time of requirement.

(2) L: This consists of a gray_scale.npy file which is the grayscale version of the MIRFLICKR25k dataset.

```python
ab_path = "../input/ab/ab/ab1.npy"
l_path = "../input/l/gray_scale.npy"
```

```python
ab_df = np.load(ab_path)
L_df = np.load(l_path)


print("AB Shape: ", ab_df.shape)
print("L Shape: ", L_df.shape)

dataset = (L_df[0:5000], ab_df[0:5000])

gc.collect()
```

```python

first_img = 0

plt.figure(figsize=(10, 10))

for i in range(1, 16, 2):
    plt.subplot(4, 4, i)
    img = np.zeros((224, 224, 3))
    img[:, :, 0] = L_df[i + first_img]
    
    img = img.astype('uint8')
    

    plt.title('Gray (X)')
    plt.axis('off')
    plt.imshow(img[:, :, 0], cmap='gray')

    plt.subplot(4, 4, i+1)
    img[:, :, 1:] = ab_df[i + first_img]

    img = cv2.cvtColor(img, cv2.COLOR_Lab2RGB)
    
    plt.title('RGB (Y)')
    plt.axis('off')
    plt.imshow(img)
```
